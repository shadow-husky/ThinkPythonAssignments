```python
# NAME: expense_tracker.py
# IMPORT STATEMENTS

#FUNCTION DEFINITIONS

# add_expense(expenses)
def add_expense(expenses):
# ask user for category
    category = input("Enter expense category: ")
# ask user for vendor
    vendor = input("Enter expense vendor: ")
# ask user for amount
    try:
        amount = float(input("Enter expense amount: "))
    except ValueError:
        print("Invalid amount. Please enter a number.")
        return
# convert amount to float
# ask user for date
    date = input("Enter expense date: ")
# create a expense dictionary
    expense = {"category": category, "vendor": vendor, "amount": amount, "date": date}
# append dictionary to expenses list
    expenses.append(expense)
    print("Expense added.")

# show_totaling_menu()
# total_all(expenses)
def total_all(expenses):
# Start total at 0
    total = 0
# loop through each expense in expenses
    for expense in expenses:
# add expense ["amount"] to total
        total += expense["amount"]
# return total
    return total

# total_by_category(expeses, category)
def total_by_category(expenses, category):
# Start total at 0
    total = 0
# loop through each expense
    for expense in expenses:
# if expense ["category"] matches vendor:
        if expense["category"].lower() == category.lower():
# add expense["amount"] to total
            total += expense["amount"]
# return total
    return total
# total by vendor
def total_by_vendor(expenses, vendor):
# start total at 0
    total = 0
# expenses vendor
    for expense in expenses:
        if expense["vendor"].lower() == vendor.lower():
# add expense amount to total
            total += expense["amount"]
    return total

# Totals menu expenses
def show_totaling_menu(expenses):
    print("\n--- Totaling Menu ---")
    print("A. Total All Expenses")
    print("C. Total by Category")
    print("V. Total by Vendor")
# Get Users category choice
    choice = input("Choice: ").lower()
# Choice A
    if choice == "a":
# Total all expenses
        total = total_all(expenses)
        print(f"Total of all expenses: ${total:.2f}")
# Totaling by category
    elif choice == "c":
        category = input("Enter category: ")
        total = total_by_category(expenses, category)
        print(f"Total of expenses in category '{category}': ${total:.2f}")
# Totaling by vendor
    elif choice == "v":
        vendor = input("Enter vendor: ")
        total = total_by_vendor(expenses, vendor)
        print(f"Total of expenses from vendor '{vendor}': ${total:.2f}")

    else:
        print("Invalid choice. Please try again.")

# PROGRAM LOGIC

# create empty expenses list
expenses = []
# while program is running
while True:
# print main menu
    print("\n--- Expense Tracker ---")
    print("1. Add Expense")
    print("2. Get Totals")
    print("3. Exit")
# ask user for selection
    selection = input("Enter your selection: ")
# if selection is 1
    if selection == "1":
# call add_expense(expenses)
        add_expense(expenses)
        print("\nCurrent expense list:")
# elif selection is 2:
    elif selection == "2":
        show_totaling_menu(expenses)
# show totaling menu
# ask user if they want all, category, or vendor total
# elif selection is 3
    elif selection == "3":
        print("Goodbye!")
# Exit Loop
        break
# else
    else:
        print("Invalid selection. Please try again.")


```
