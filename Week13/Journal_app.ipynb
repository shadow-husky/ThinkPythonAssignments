```python
# NAME: journal.py
# IMPORT STATEMENTS

# FUNCTION DEFINITIONS
# write_entry(filename, entry)
def write_entry(filename, entry):
# Open file in append mode ('a')
    f = open(filename, 'a')
# Write entry + newline (\n) so each entry is on its own line
    f.write(entry + "\n")
# close file
    f.close()
    print("Entry Saved.")

# read_journal(filename)
# Open file in read mode ('r')
# Read entire contents
# close file
# FileNotFoundError print message saying no journal exists

# PROGRAM LOGIC

# set filename = "journal.txt"
filename = "journal.txt"

# Loop forever
while True:
# ask user for write, read, or quit
  choice = input("Do you want to (W)rite, (R)ead, or (Q)uit? ").lower()
# IF W ask for entry
  if choice == 'w':
      entry = input("Enter your journal entry: ")
# Call write_entry
      write_entry(filename, entry)
# IF R call read_journal
  elif choice == 'r':
      print("Feature coming soon")
# IF Q Loop end
  elif choice == 'q':
      print("Goodbye")
      break
# Print invalid input
  else:
      print("Invalid input. Please try again.")

```
