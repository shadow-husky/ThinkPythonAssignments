```python
# NAME: gradebook.py

# IMPORT STATEMENTS
# (None)

# FUNCTION DEFINITIONS

def add_students(gradebook):
# While loop to keep asking for student names
    while True:
      name = input("Enter student name (or 'done'): ")
# If user types "done", end loop
      if name.lower() == "done":
          break
# Otherwise
# Ask for the student's grade
      grade = int(input(f"Enter {name}'s grade: "))
# Convert grade to integer
      gradebook[name] = grade
# Store it in dictionary

def calc_average(gradebook):
# Handle case where dictionary is empty to avoid division by zero
    if len(gradebook) == 0:
        return 0
# Get all grades using gradebook.values()
# Add them together
    total = sum(gradebook.values())
# Divide by the number of students
    average = total / len(gradebook)
# Return the average
    return average

# PROGRAM LOGIC
# Create empty dictionary called gradebook
gradebook = {}
# Call add_students to fill the dictionary
add_students(gradebook)
# Call calc_average to get average
average = calc_average(gradebook)
# Print each student and their grade
print("\n--- Class Report ---")
for name, grade in gradebook.items():
    print(f"{name}: {grade}")
# Print class average formatted to 2 decimal places
print(f"Class Average: {average:.2f}")
```
