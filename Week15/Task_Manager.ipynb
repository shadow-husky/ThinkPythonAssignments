```python
COMPLETED_FILE = "completed_tasks.txt"


def get_priority_label(priority):
    return {
        1: "High",
        2: "Medium",
        3: "Low"
    }.get(priority, "Unknown")


def add_task(tasks):
    name = input("Enter task name: ").strip()

    while True:
        try:
            priority = int(input("Enter priority (1 = High, 2 = Medium, 3 = Low): "))
            if priority in [1, 2, 3]:
                break
            print("Please enter 1, 2, or 3.")
        except ValueError:
            print("Invalid input. Please enter a number.")

    tasks.append({"name": name, "priority": priority})
    print(f"Task '{name}' added.\n")


def view_tasks(tasks):
    if not tasks:
        print("No tasks available.\n")
        return []

    sorted_tasks = sorted(tasks, key=lambda x: x["priority"])

    print("\n--- Your Tasks ---")
    for i, task in enumerate(sorted_tasks, start=1):
        label = get_priority_label(task["priority"])
        print(f"{i}. {task['name']} (Priority: {label})")
    print()

    return sorted_tasks


def save_completed_task(task_name):
    with open(COMPLETED_FILE, "a") as file:
        file.write(task_name + "\n")


def complete_task(tasks):
    if not tasks:
        print("No tasks to complete.\n")
        return

    sorted_tasks = view_tasks(tasks)

    try:
        choice = int(input("Enter the number of the task to complete: "))
        if choice < 1 or choice > len(sorted_tasks):
            print("Invalid task number.\n")
            return

        task_to_complete = sorted_tasks[choice - 1]
        tasks.remove(task_to_complete)

        save_completed_task(task_to_complete["name"])

        print(f"Completed task: {task_to_complete['name']}")
        print("Saved to completed task history.\n")

    except ValueError:
        print("Please enter a valid number.\n")
def view_completed_tasks():
    try:
        with open(COMPLETED_FILE, "r") as file:
            lines = file.readlines()

        if not lines:
            print("No completed tasks yet.\n")
            return

        print("\n--- Completed Tasks ---")
        for i, line in enumerate(lines, start=1):
            print(f"{i}. {line.strip()}")
        print()

    except FileNotFoundError:
        print("No completed tasks file found yet.\n")

def main():
    tasks = []

    while True:
        print("=== Personal Task Manager ===")
        print("1. Add Task")
        print("2. View Tasks")
        print("3. Complete Task")
        print("4. View Completed Tasks")
        print("5. Exit")

        choice = input("Choose an option: ").strip()

        if choice == "1":
            add_task(tasks)
        elif choice == "2":
            view_tasks(tasks)
        elif choice == "3":
            complete_task(tasks)
        elif choice == "4":
            view_completed_tasks()
        elif choice == "5":
            print("Exiting the program. Goodbye!")
            break
        else:
            print("Invalid choice. Try again.\n")


if __name__ == "__main__":
    main()
```
