```python
import json
import os

SAVE_FILE = "games.json"

VALID_STATUSES = {
    "1": "Not Started",
    "2": "Playing",
    "3": "Completed"
}


def load_games():
    if not os.path.exists(SAVE_FILE):
        with open(SAVE_FILE, "w") as file:
            json.dump([], file)
        return []

    with open(SAVE_FILE, "r") as file:
        return json.load(file)


def save_games(games):
    with open(SAVE_FILE, "w") as file:
        json.dump(games, file, indent=4)


def add_game(games):
    title = input("Game title: ").strip()
    platform = input("Platform: ").strip()
    genre = input("Genre: ").strip()

    while True:
        try:
            hours_played = float(input("Hours played: "))
            if hours_played >= 0:
                break
            print("Hours cannot be negative.")
        except ValueError:
            print("Please enter a valid number.")

    print("\nStatus:")
    print("1. Not Started")
    print("2. Playing")
    print("3. Completed")

    while True:
        status_choice = input("Choose status: ").strip()
        if status_choice in VALID_STATUSES:
            status = VALID_STATUSES[status_choice]
            break
        print("Please choose 1, 2, or 3.")

    games.append({
        "title": title,
        "platform": platform,
        "genre": genre,
        "hours_played": hours_played,
        "status": status
    })

    save_games(games)
    print(f"\nAdded '{title}' to your backlog.\n")


def view_games(games):
    if not games:
        print("\nNo games saved yet.\n")
        return

    print("\n=== Your Game Backlog ===")
    for index, game in enumerate(games, start=1):
        print(f"{index}. {game['title']}")
        print(f"   Platform: {game['platform']}")
        print(f"   Genre: {game['genre']}")
        print(f"   Hours Played: {game['hours_played']}")
        print(f"   Status: {game['status']}")
        print()


def view_stats(games):
    if not games:
        print("\nNo games saved yet.\n")
        return

    total_games = len(games)
    total_hours = sum(game["hours_played"] for game in games)
    completed = sum(1 for game in games if game["status"] == "Completed")
    unfinished = total_games - completed

    print("\n=== Backlog Stats ===")
    print(f"Total Games: {total_games}")
    print(f"Total Hours Played: {total_hours:.1f}")
    print(f"Completed Games: {completed}")
    print(f"Unfinished Games: {unfinished}")
    print()


def main():
    games = load_games()

    while True:
        print("=== Game Backlog Tracker ===")
        print("1. Add Game")
        print("2. View Games")
        print("3. View Stats")
        print("4. Exit")

        choice = input("Choose an option: ").strip()

        if choice == "1":
            add_game(games)
        elif choice == "2":
            view_games(games)
        elif choice == "3":
            view_stats(games)
        elif choice == "4":
            print("Goodbye. Your backlog is safely saved 🎮")
            break
        else:
            print("Invalid option. Try again.\n")


if __name__ == "__main__":
    main()
```
