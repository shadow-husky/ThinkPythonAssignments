```python
# NAME: rectangle.py

# IMPORT STATEMENTS

# FUNCTION DEFINITIONS
# calculate rectangle length and width
def calc_rectangle(length, width):
# calculate the area using length * width
    area = length * width
# Put both values into tuple (area and perimeter)
    perimeter = 2 * (length + width)
# return the tuple
    return (area, perimeter)

# PROGRAM LOGIC
# ask user for length
length = float(input("Enter length: "))
# ask user for width
width = float(input("Enter width: "))
# call the function and store the results
# unpack tuple into two variables
# my_area, my_perimeter = result
my_area, my_perimeter = calc_rectangle(length, width)

# print results
print("\n--- Results ---")
# print area
print("The Area is:", my_area)
# print perimeter
print("The Perimeter is:", my_perimeter)
```
